```python
import requests
from bs4 import BeautifulSoup

def get_response():
    url = "https://packages.ubuntu.com/jammy/allpackages"
    response = requests.get(url)
    return response



def get_package_license(package_name):
    url = f"https://packages.ubuntu.com/jammy/{package_name}"
    response = requests.get(url)
    soup = BeautifulSoup(response.text, 'html.parser')
    print(soup)
    license_element = soup.find('th', text='License:')
    if license_element:
        license_info = license_element.find_next('td').text.strip()
    else:
        license_info = "Unknown"
    return license_info

```

```python
import subprocess
import json

def get_multiverse_packages():
    command = "winget --list --exact"
    result = subprocess.check_output(command, shell=True).decode()
    packages = json.loads(result)
    multiverse_packages = []

    for package in packages:
        if "multiverse" in package["Categories"]:
            multiverse_packages.append(package["Name"])

    return multiverse_packages

if __name__ == "__main__":
    multiverse_packages = get_multiverse_packages()
    print("Multiverse Packages:")
    for package in multiverse_packages:
        print(package)
```
